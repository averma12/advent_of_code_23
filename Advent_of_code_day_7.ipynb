```python
def parse_input(input_text):
    # Split the input into lines
    lines = input_text.strip().split('\n')

    # Convert each line into a tuple (hand, bid)
    hands_and_bids = []
    for line in lines:
        # Split line into hand and bid
        hand, bid = line.split()
        # Convert bid to integer
        bid = int(bid)
        # Add tuple to list
        hands_and_bids.append((hand, bid))

    return hands_and_bids

# Test the parsing
input_text = """32T3K 765
T55J5 684
KK677 28
KTJJT 220
QQQJA 483"""

hands_and_bids = parse_input(input_text)
print(hands_and_bids)
```

```python
def get_hand_type(hand):
    # Count non-joker cards
    counts = Counter(card for card in hand if card != 'J')
    joker_count = hand.count('J')

    # If all jokers, it's five of a kind
    if joker_count == 5:
        return 7

    # Add jokers to the highest count
    freq_values = sorted(counts.values(), reverse=True) if counts else [0]
    # Add jokers to most frequent card
    freq_values[0] += joker_count

    highest_freq = freq_values[0]

    if highest_freq == 5:
        return 7  # Five of a kind
    elif highest_freq == 4:
        return 6  # Four of a kind
    elif highest_freq == 3:
        if len(freq_values) > 1 and freq_values[1] == 2:
            return 5  # Full House
        return 4    # Three of a Kind
    elif highest_freq == 2:
        if len(freq_values) > 1 and freq_values[1] == 2:
            return 3  # Two Pair
        return 2    # One Pair
    return 1  # High card

def get_hand_strength(hand_and_bid):
    hand = hand_and_bid[0]
    type_strength = get_hand_type(hand)

    # Updated card ranks with J as lowest
    card_ranks = ['A', 'K', 'Q', 'T', '9', '8', '7', '6', '5', '4', '3', '2', 'J']
    card_values = {card: 13 - i for i, card in enumerate(card_ranks)}

    card_strengths = [card_values[card] for card in hand]

    return (type_strength, card_strengths)

# Test hands
input_text = """32T3K 765
T55J5 684
KK677 28
KTJJT 220
QQQJA 483"""

hands_and_bids = [(hand, int(bid)) for hand, bid in (line.split() for line in input_text.splitlines())]

# Let's see each hand's type and strength
for hand, bid in hands_and_bids:
    hand_type = get_hand_type(hand)
    print(f"Hand: {hand}, Type: {hand_type}")

print("\nSorted hands:")
sorted_hands = sorted(hands_and_bids, key=get_hand_strength)
for rank, (hand, bid) in enumerate(sorted_hands, 1):
    print(f"Rank {rank}: {hand} (Type: {get_hand_type(hand)}) Bid: {bid}")

total = sum(rank * bid for rank, (hand, bid) in enumerate(sorted_hands, 1))
print(f"\nTotal winnings: {total}")
```
