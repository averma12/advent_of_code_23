```python
def parse_input(input_text):
    # Split into lines and get instructions and node definitions
    lines = input_text.splitlines()

    # First line is the instructions
    instructions = lines[0]

    # Initialize empty graph
    graph = {}

    # Process remaining lines for node definitions
    for line in lines[1:]:
        # Skip empty lines if any
        if not line.strip():
            continue

        # Split into node and connections
        node, connections = line.split("=")
        node = node.strip()

        # Clean up the connections string and split into left/right
        # Remove parentheses and split on comma
        connections = connections.strip()[1:-1].split(",")
        # Clean up whitespace from each connection
        connections = [x.strip() for x in connections]

        # Add to graph
        graph[node] = connections

    return instructions, graph
```

```python
def find_path(graph, instructions, start='AAA', target='ZZZ'):
    steps = 0               # Counter for steps taken
    current_node = start    # Track where we are

    while current_node != target:
        # Get current instruction using modulo to cycle through instructions
        current_instruction = instructions[steps % len(instructions)]

        # Choose path based on instruction
        if current_instruction == "L":
            current_node = graph[current_node][0]
        else:
            current_node = graph[current_node][1]

        steps += 1

    return steps
```

```python
def follow_paths(graph, instructions):
    # Get our starting positions
    current_positions = find_start_nodes(graph)
    print(f"Starting positions: {current_positions}")  # For debugging
    steps = 0

    while True:
        # Get current instruction
        instruction = instructions[steps % len(instructions)]

        # Move each position based on instruction
        next_positions = []
        for current in current_positions:
            # Choose left (0) or right (1) path
            path_index = 0 if instruction == 'L' else 1
            next_position = graph[current][path_index]
            next_positions.append(next_position)

        steps += 1
        current_positions = next_positions

        # Debug print to see movement
        print(f"Step {steps}: {current_positions}")

        # Check if all current positions end in Z
        if all(pos.endswith('Z') for pos in current_positions):
            return steps
```

```python
def find_cycle_length(graph, instructions, start_pos):
    # Keep track of positions we've seen and when we saw them
    seen = {}  # Will store (position, instruction_index) -> step_number

    current_pos = start_pos
    steps = 0

    while True:
        # Get current instruction index (our position in instruction string)
        instruction_idx = steps % len(instructions)
        instruction = instructions[instruction_idx]

        # Create a state tuple (where we are, what instruction we're at)
        current_state = (current_pos, instruction_idx)

        # If we've seen this state before, we found a cycle!
        if current_state in seen:
            cycle_start = seen[current_state]
            return steps - cycle_start

        # Record this state
        seen[current_state] = steps

        # Move to next position
        path_idx = 0 if instruction == 'L' else 1
        current_pos = graph[current_pos][path_idx]

        steps += 1
```

```python
from functools import reduce
from math import gcd

def lcm(a, b):
    return abs(a * b) // gcd(a, b)

def find_lcm_of_list(numbers):
    return reduce(lcm, numbers)

# Example 1: Simple addition
numbers = [1, 2, 3, 4]
sum = reduce(lambda x, y: x + y, numbers)
# Here's what happens step by step:
# 1. Start with 1,2 -> 3
# 2. Take 3,3 -> 6
# 3. Take 6,4 -> 10
print(sum)  # 10

# Example 2: Finding maximum
max_num = reduce(lambda x, y: x if x > y else y, numbers)
# 1. Compare 1,2 -> 2
# 2. Compare 2,3 -> 3
# 3. Compare 3,4 -> 4
print(max_num)  # 4

# In our LCM case:
numbers = [4, 6, 8]
result = reduce(lcm, numbers)
print(result)
# 1. lcm(4,6) = 12
# 2. lcm(12,8) = 24
```

```python
def solve_ghost_paths(graph, instructions):
    # Find all starting positions (nodes ending in A)
    start_positions = [node for node in graph.keys() if node.endswith('A')]
    print(f"Starting positions: {start_positions}")

    # Find cycle length for each starting position
    cycle_lengths = []
    for start in start_positions:
        length = find_cycle_length(graph, instructions, start)
        cycle_lengths.append(length)
        print(f"Path from {start} has cycle length: {length}")

    # Calculate LCM of all cycle lengths
    result = find_lcm_of_list(cycle_lengths)
    return result
```

```python
solve_ghost_paths(graph, instructions)
```
