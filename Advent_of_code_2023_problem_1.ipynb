```python
def extract_first_last_digit(s):
    digits = [char for char in s if char.isdigit()]
    if not digits:
        return 0

    return int(digits[0] + digits[-1])
```

```python
def extract_first_last_digit(s):
    digits = [char for char in s if char.isdigit()]
    if not digits:
        return None
    elif len(digits) == 1:
        return digits[0]
    else:
        return digits[0] + digits[-1]
```

```python
def extract_first_last_digit_iterator(s):
    first_digit = next((char for char in s if char.isdigit()), None)
    last_digit = next((char for char in reversed(s) if char.isdigit()), None)
    if not first_digit:
        return 0
    return int(first_digit + last_digit)

```

```python
def extract_digits(s):
    digits_dict = {
        "zero": '0',
        "one": '1',
        "two": '2',
        "three": '3',
        "four": '4',
        "five": '5',
        "six": '6',
        "seven": '7',
        "eight": '8',
        "nine": '9'
    }
    found_digits = []
    i = 0
    while i < len(s):
        # First, check if the character is a digit
        if s[i].isdigit():
            found_digits.append(s[i])
            i += 1
            continue
        # Then, check for spelled-out numbers starting at position i
        matched = False
        for word in digits_dict:
            word_len = len(word)
            if s[i:i+word_len] == word:
                found_digits.append(digits_dict[word])
                i += word_len  # Move index forward by the length of the matched word
                matched = True
                break  # Exit the loop after a match
        if not matched:
            i += 1  # Move to the next character
    # Now extract first and last digits
    if not found_digits:
        return 0
    else:
        return int(found_digits[0] + found_digits[-1])
```

```python
def extract_digits_alternate(s):
    digits_dict = {
        "zero": '0',
        "one": '1',
        "two": '2',
        "three": '3',
        "four": '4',
        "five": '5',
        "six": '6',
        "seven": '7',
        "eight": '8',
        "nine": '9'
    }
    found_digits = []
    i = 0
    while i < len(s):
        # First, check if the character is a digit
        if s[i].isdigit():
            found_digits.append(s[i])
            i += 1
            continue
        # Then, check for spelled-out numbers starting at position i
        matched = False
        for word in digits_dict:
            word_len = len(word)
            if s[i:i+word_len] == word:
                found_digits.append(digits_dict[word])
                i += word_len - 1  # Move index forward by the length of the matched word
                matched = True
                break  # Exit the loop after a match
        if not matched:
            i += 1  # Move to the next character
    # Now extract first and last digits
    if not found_digits:
        return 0
    else:
        return int(found_digits[0] + found_digits[-1])
```
