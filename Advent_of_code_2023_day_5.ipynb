```python
import re
mappings = {}

lines = sample.splitlines()
pattern = r'(\w+)-to-(\w+) map:'

for i, line in enumerate(lines):
    match = re.search(pattern, line)
    if match:
        source = match.group(1)
        destination = match.group(2)

        # Find all number lines that follow until empty line or next mapping
        numbers = []
        current = i + 1
        while current < len(lines) and lines[current].strip():
            numbers.append(lines[current])
            current += 1

        # Store in dictionary with key like 'seed-to-soil'
        mapping_name = f'{source}-to-{destination}'
        mappings[mapping_name] = numbers
```

```python
def create_ranges_for_line(line):
    # Convert "50 98 2" into three integers [50, 98, 2]
    dest_start, source_start, length = map(int, line.split())

    # Create the ranges
    source_range = range(source_start, source_start + length)
    dest_range = range(dest_start, dest_start + length)

    return source_range, dest_range

# Let's test it with the first line from seed-to-soil
test_line = "50 98 2"
source_range, dest_range = create_ranges_for_line(test_line)

print(f"Source range: {list(source_range)}")  # Should print [98, 99]
print(f"Dest range: {list(dest_range)}")    # Should print [50, 51]
```

```python
def create_mapping_ranges(mappings):
    all_mappings = {}
    for mapping_name, mapping_lines in mappings.items():
        source_range = []
        dest_range = []
        for line in mapping_lines:
            source_range_line, dest_range_line = create_ranges_for_line(line)
            source_range.extend(list(source_range_line))
            dest_range.extend(list(dest_range_line))
        all_mappings[mapping_name] = (source_range, dest_range)
    return all_mappings
```

```python
def get_column_names(mapping_name):
    # From "seed-to-soil" get ["seed", "soil"]
    pattern = r'(\w+)-to-(\w+)'
    match = re.search(pattern, mapping_name)
    return match.group(1), match.group(2)
```

```python
import pandas as pd
```

```python
def create_mapping_dataframes(all_mapping_ranges):
    dataframes = {}
    for mapping_name, (source_range, dest_range) in all_mapping_ranges.items():
        source_col, dest_col = get_column_names(mapping_name)
        df = pd.DataFrame({
            source_col: source_range,
            dest_col: dest_range
        }).reset_index(drop=True)  # Drop the index completely
        dataframes[mapping_name] = df
    return dataframes
```

```python
def process_mapping_chain(all_dfs, seed_values=None):
    """
    Process the complete chain of mappings from seed to location.
    """
    seed_to_soil = all_dfs['seed-to-soil']
    soil_to_fert = all_dfs['soil-to-fertilizer']
    # First merge
    first_merge = pd.merge(seed_to_soil, soil_to_fert, on='soil', how='outer').fillna(-1)

    # Fix seed column: if seed is -1, use soil value
    first_merge['seed'] = first_merge.apply(lambda row: row['soil'] if row['seed'] == -1 else row['seed'], axis=1)

    # Second merge
    second_merge = pd.merge(first_merge, fert_to_water, on='fertilizer', how='outer').fillna(-1)
    # Fix previous columns similarly

    # Fix in order from left to right:
    second_merge['fertilizer'] = second_merge.apply(lambda row: row['soil'] if row['fertilizer'] == -1 else row['fertilizer'], axis=1)
    second_merge['water'] = second_merge.apply(lambda row: row['fertilizer'] if row['water'] == -1 else row['water'], axis=1)
    second_merge['soil'] = second_merge.apply(lambda row: row['fertilizer'] if row['soil'] == -1 else row['soil'], axis=1)
    second_merge['seed'] = second_merge.apply(lambda row: row['soil'] if row['seed'] == -1 else row['seed'], axis=1)
    # Third merge
    water_to_light = all_dfs['water-to-light']
    third_merge = pd.merge(second_merge, water_to_light, on='water', how='outer').fillna(-1)

    # Fix any -1s in the new columns from left to right
    third_merge['light'] = third_merge.apply(lambda row: row['water'] if row['light'] == -1 else row['light'], axis=1)
    # Fourth merge
    light_to_temp = all_dfs['light-to-temperature']
    fourth_merge = pd.merge(third_merge, light_to_temp, on='light', how='outer').fillna(-1)
    fourth_merge['temperature'] = fourth_merge.apply(lambda row: row['light'] if row['temperature'] == -1 else row['temperature'], axis=1)

    # Fifth merge
    temp_to_humidity = all_dfs['temperature-to-humidity']
    fifth_merge = pd.merge(fourth_merge, temp_to_humidity, on='temperature', how='outer').fillna(-1)
    fifth_merge['humidity'] = fifth_merge.apply(lambda row: row['temperature'] if row['humidity'] == -1 else row['humidity'], axis=1)

    #Final merge
    humidity_to_location = all_dfs['humidity-to-location']
    final_merge = pd.merge(fifth_merge, humidity_to_location, on='humidity', how='outer').fillna(-1)
    final_merge['location'] = final_merge.apply(lambda row: row['humidity'] if row['location'] == -1 else row['location'], axis=1)

    return final_merge

```

```python
def get_all_mappings(number, mappings):
    """
    Get all mappings for a number through the complete chain.

    Args:
        number: Starting number (seed)
        mappings: Dictionary containing all mapping rules

    Returns:
        Dictionary with all mapped values
    """
    result = {'seed': number}

    # Order of mappings
    mapping_chain = [
        'seed-to-soil',
        'soil-to-fertilizer',
        'fertilizer-to-water',
        'water-to-light',
        'light-to-temperature',
        'temperature-to-humidity',
        'humidity-to-location'
    ]

    current = number
    for mapping_name in mapping_chain:
        current = find_mapping(current, mappings[mapping_name])
        # Get the destination name from the mapping (e.g., 'soil' from 'seed-to-soil')
        dest_name = mapping_name.split('-to-')[1]
        result[dest_name] = current

    return result

# Usage example:
# seed = 79
# mapped_values = get_all_mappings(seed, mappings)
# This would return something like:
# {
#     'seed': 79,
#     'soil': 81,
#     'fertilizer': 81,
#     'water': 81,
#     'light': 74,
#     'temperature': 78,
#     'humidity': 78,
#     'location': 82
# }
```
